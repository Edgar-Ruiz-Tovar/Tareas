# Puntos Fijos

```python
import math
import matplotlib.pyplot as plt
import numpy as np
    
def punto_fijo(x,iteraciones,funcion):
    for i in range(iteraciones+1):
        print("x{}={}".format(i,x))
        x=funcion(x)
        
def lista_punto_fijo(x,iteraciones,funcion):
    lista=[]
    for i in range(iteraciones+1):
        lista.append(x)
        x=funcion(x)
    return lista

def grafica(a,b,inicio,funcion):
    f2 = np.vectorize(funcion) # porque np.int no acepta un array
    p=lista_punto_fijo(inicio,10,f2)
    abcisa=np.arange(a,b,0.01)
    ordenada=f2(abcisa)
    plt.plot(abcisa,ordenada,linewidth=3,label='Función g')
    plt.scatter(p[:-1],p[1:],color='green',linewidth=2,label='Iteraciones')
    plt.xlabel('x')
    plt.ylabel('g(x)')
    plt.grid(True)
    plt.legend()
    plt.show()
```

### Algoritmo babilónico como iteración de punto fijo
Sea $A > 0$ un número dado. Vamos a aplicar el método de punto fijo a la función
$$g(x)=\dfrac{1}{2}\left(x+\dfrac{A}{x}\right)$$
* Demuestre que $g$ cumple con las condiciones del teorema sobre la convergencia de punto fijo en cualquier intervalo $[\sqrt{A},B]$ donde $B>\sqrt{A}$.

Veamos que
$$g^\prime(x)=\dfrac{1}{2}\left(1-\dfrac{A}{x^2}\right)$$
$$g^{\prime\prime}(x)=\dfrac{1}{2}\left(\dfrac{2A}{x^3}\right)=\dfrac{A}{x^3}>0\quad\forall x\in[\sqrt{A},B]$$
luego la derivada es creciento y por lo tanto
$$\min |g^\prime(x)|=g^\prime(\sqrt{A})=\dfrac{1}{2}\left(1-\dfrac{A}{(\sqrt{A})^2}\right)=\dfrac{1}{2}\left(1-\dfrac{A}{A}\right)=0$$
$$\max |g^\prime(x)|=g^\prime(B)=\dfrac{1}{2}\left(1-\dfrac{A}{B^2}\right)=\dfrac{1}{2}\left(1-\dfrac{A}{B^2}\right)$$
Por lo tanto
$$C=\max |g^\prime(x)|=\max\left\lbrace 0,\dfrac{1}{2}\left(1-\dfrac{A}{B^2}\right)\right\rbrace=\dfrac{1}{2}\left(1-\dfrac{A}{B^2}\right)<1$$
La última desigualdad se sigue de que
$$\dfrac{A}{B^2}>0$$
$$\dfrac{A}{B^2}+1>0$$
$$-\dfrac{A}{B^2}<1$$
$$1-\dfrac{A}{B^2}<2$$
$$\dfrac{1}{2}\left(1-\dfrac{A}{B^2}\right)<1$$
Por otro lado
$$g(\sqrt{A})=\dfrac{1}{2}\left(\sqrt{A}+\dfrac{A}{\sqrt{A}}\right)=\dfrac{1}{2}\left(\sqrt{A}+\sqrt{A}\right)=\dfrac{2\sqrt{A}}{2}=\sqrt{A}$$
$$g(B)=\dfrac{1}{2}\left(B+\dfrac{A}{B}\right)<B$$
La última desigualdad se sigue de
$$\sqrt{A}<B$$
$$A<B^2$$
$$B^2+A<2B^2$$
$$\dfrac{B^2+A}{B}<2B$$
$$B+\dfrac{A}{B}<2B$$
$$\dfrac{1}{2}\left(B+\dfrac{A}{B}\right)<B$$
Por lo tanto $g$ es contractiva en $[\sqrt{A},B]_\blacksquare$
* Muestre que $\sqrt{A}$ es el único punto fijo positivo de $g$.
Como $g$ es contractiva en $[\sqrt{A},B]$ y
$$g(\sqrt{A})=\dfrac{1}{2}\left(\sqrt{A}+\dfrac{A}{\sqrt{A}}\right)=\dfrac{1}{2}\left(\sqrt{A}+\sqrt{A}\right)=\dfrac{2\sqrt{A}}{2}=\sqrt{A}$$
entonces por la unicidad del punto fijo en las funciones contractivas se concluye que $\sqrt{A}$ es el único punto fijo$_\blacksquare$
* Muestre que $\dfrac{1+A}{2}\geq \sqrt{A}$ para todo $A > 0$. Por lo tanto, el punto $x_0=\dfrac{1+A}{2}$ se puede usar como punto inicial.
$$(1-\sqrt{A})^2\geq 0$$
$$1-2\sqrt{A}+A\geq 0$$
$$1+A\geq 2\sqrt{A}$$
$$\dfrac{1+A}{2}\geq \sqrt{A}$$

## Puntos Fijos de Funciones Contractivas

### Ejemplo 10
$$g(x)=\dfrac{x^2-1}{3}\quad\text{en }[-1,1]$$
Notemos que
$$g^\prime(x)=\dfrac{2}{3}x$$
Como $g$ es una parábola entonces tenemos que es simétrica respecto al eje $y$ y por tanto el valor máximo en el intervalo [-1,1] lo podremos calcular como
$$C=|g^\prime(x)|=|g^\prime(1)|=\dfrac{2}{3}(1)=\dfrac{2}{3}<1$$
Además
$$g(-1)=\dfrac{2}{3}(-1)=-\dfrac{2}{3}$$
$$g(1)=\dfrac{2}{3}(1)=\dfrac{2}{3}$$
Por lo tanto concluimos que $g$ es contractiva en [-1,1]. Entonces tiene un punto fijo y viene dado por

```python
def ejer10(x):
    return (x**2-1)/3
punto_fijo(0,10,ejer10)
```

El punto fijo vemos que es aproximadamente -0.3027. Comprobando
$$g(-0.3027)=\dfrac{(-0.3027)^2-1}{3}=-0.3027$$
Y graficando

```python
grafica(-1.0,1.0,0,ejer10)
```

### Ejemplo 13
Verificar las hipótesis del teorema para la función
$$g(x)=\pi+\dfrac{1}{2}\sin\left(\dfrac{x}{2}\right)\quad\text{en }[0,2\pi]$$
Tenemos que
$$g^\prime(x)=\dfrac{1}{4}\cos\left(\dfrac{x}{2}\right)$$
Como $-1\leq \cos(x)\leq 1$ para toda $x$ entonces 
$$-\dfrac{1}{4}\leq g^\prime(x)\leq \dfrac{1}{4}$$
Y por lo tanto 
$$C=\max |g^\prime(x)|=\dfrac{1}{4}<1$$
Note que además
$$\pi-\dfrac{1}{2}\leq\pi+\dfrac{1}{2}\sin\left(\dfrac{x}{2}\right)\leq\dfrac{1}{2}+\pi$$
por lo que $g([0,2\pi])\subseteq [0,2\pi]$. Por lo tanto $g$ es una contracción. Ahora encontremos por cuál es por medio de itereaciones.

```python
def ejer13(x):
    return math.pi+math.sin(math.radians(x)/2)
punto_fijo(0.4,10,ejer13)
```

Comprobando

```python
print(ejer13(3.169246016815674))
```

### Ejercicio 4
$$g(x)=2+\dfrac{x}{2}-\dfrac{x^2}{4}\quad\text{en}\;[1,3]$$
Su derivada es
$$g^\prime(x)=\dfrac{1}{2}-\dfrac{x}{2}$$
Con lo que vemos que $g$ es decreciente en [1,3] por lo que
$$C=\max |g^\prime(x)|=0<1$$
Además de que
$$g(1)=2+\dfrac{1}{2}-\dfrac{1^2}{4}=2.25$$
$$g(3)=2+\dfrac{3}{2}-\dfrac{3^2}{4}=2.75$$
Por lo tanto $g$ es contractiva en [1,3] y entonces tiene un único punto fijo, el cual viene dado por

```python
def ejer4(x):
    return 2+x/2-x**2/4

punto_fijo(1,10,ejer4)
```

Lo que nos dice que el punto fijo es en 2. Comprobando:
$$g(2)=2+\dfrac{2}{2}-\dfrac{2^2}{4}=2+1-1=2$$
Por útlimo veamos una gráfica de la función $g$ en el intervalo [1,3] y note cómo las iteraciones poco a poco se van acercando al punto fijo.

```python
grafica(1.03,3.0,1,ejer4)
```

### Ejercicio 6
$$g(x)=\dfrac{1}{2\sqrt{2x}}\quad\text{en}\;[0.4,1]$$
Su derivada es
$$g^\prime(x)=\dfrac{1}{2\sqrt{2x}}$$
Con lo que vemos que $g$ es decreciente en [0.4,1] por lo que
$$C=\max |g^\prime(x)|=g^\prime(0.4)=0.559<1$$
Además de que
$$g(1)=\dfrac{1}{2\sqrt{2(0.04)}}=0.559$$
$$g(3)=\dfrac{1}{2\sqrt{(2)(1)}}=.35$$
Por lo tanto $g$ es contractiva en [0.4,1] y entonces tiene un único punto fijo, el cual viene dado por

```python
def ejer6(x):
    return 1/(2*math.sqrt(2*x))

punto_fijo(1,10,ejer6)
```

Lo que nos dice que el punto fijo es en 0.5. Comprobando:
$$g(0.5)=\dfrac{1}{2\sqrt{2(0.5)}}=\dfrac{1}{2\sqrt{1}}=\dfrac{1}{2}=0.5$$
Veamos la gráfica

```python
grafica(0.4,1.0,1,ejer6)
```

Si vemos la misma función en el intervalo [3,4] tenemos que
$$g(3)=\dfrac{3^2-1}{3}=\dfrac{8}{3}$$
$$g(4)=\dfrac{4^2-1}{3}=\dfrac{15}{3}=5$$
Como $g(4)>4$ entonces $g$ no es contractiva.

### Ejemplo 19
$$x-\cos x=0$$
Sea $f(x)=x-\cos x$ en $[0,\pi]$. Luego, si $p$ es una raíz de $f$ entonces $p$ será un punto fijo de
$$g(x)=x-f(x)=x-(x-\cos x)=\cos x$$
Note que que $g$ es contractiva en $[0,\pi]$ pues 
$$g^\prime(x)=-\sin x$$
Y como $|\sin x|\leq 1$ entonces
$$C=\max |g^\prime(x)|=1$$

```python
def ejer19(x):
    return math.cos(math.radians(x))
punto_fijo(0.0,10,ejer19)
grafica(0.0,math.pi,0.0,ejer19)
```

```python
print(ejer19(0.9998477415310881))
```
